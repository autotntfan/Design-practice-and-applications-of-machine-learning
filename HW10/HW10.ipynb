```python
titan=pd.read_csv('titanic_train.csv')
truthtable=titan.isnull().any()
for i,data in enumerate(truthtable):
    if data==True:
        print('null',titan.columns.values[i])
```

```python
for i,data in enumerate(titan.dtypes):
    if(data == "object"):
        print(titan.columns.values[i],"此欄位數據格式非數字")
    else:
        continue
```

```python

age=sns.FacetGrid(titan,col='survived')
age.map(sns.distplot,'age',kde=True)
```

```python
sns.countplot(titan['pclass'], hue=titan['survived'])
```

```python
#print("----------------填補 age 的缺失值--------------------")    
print("把名子中的稱謂合併剩下四種 分別為 Mr、Dr、Miss、Mrs、Master，而age的缺失值是利用這四種稱謂的平均值而填補")
titan['Title1'] = titan['name'].str.split(", ",expand=True)[1]
titan['Title1'] = titan['Title1'].str.split(". ",expand=True)[0]
print(titan['Title1'].unique())

pd.crosstab(titan['Title1'],titan['sex']).T.style.background_gradient(cmap='summer_r')
# pd.crosstab(data['Title1'],data['survived']).T.style.background_gradient(cmap='summer_r')
```

```python
titan['Title2'] = titan['Title1'].replace(['Mlle','Mme','Ms','Major','Lady','th','Don','Col','Capt','Dona','Jonkheer','Rev','Sir'],
                                        ['Miss','Miss','Miss','Dr','Mrs','Mrs','Mr','Dr','Dr','Mrs','Mr','Mr','Dr'])

pd.crosstab(titan['Title2'],titan['sex']).T.style.background_gradient(cmap='summer_r')
```

```python
data=titan
mask =  (data['age'].isnull()) & ( (data['name'].str.contains("Mrs")) | (data['name'].str.contains("Miss") |
                                                                         (data['name'].str.contains("Ms"))) )
mask2 = ( (data['name'].str.contains("Mrs")) | (data['name'].str.contains("Miss")) |
                                                                         (data['name'].str.contains("Ms")))
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())

mask =  (data['age'].isnull()) & ( (data['name'].str.contains("Mr")) | (data['name'].str.contains("Dr")))
mask2 = ( (data['name'].str.contains("Mr")) | (data['name'].str.contains("Dr")) )
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())  

mask =  (data['age'].isnull()) & ( data['name'].str.contains("Master"))
mask2 = ( (data['name'].str.contains("Master")))
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())  

data.drop(['name','Title1'],axis=1,inplace = True) 
print("--------------------age已無缺失值--------------------")
Truthtable = data.isnull().any().tolist()

for i in range(len(Truthtable)):
    if(Truthtable[i] == True):
        print(data.columns.values[i],"此數據有缺失值")
```
