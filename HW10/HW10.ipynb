```python
#print("----------------填補 age 的缺失值--------------------")    
print("把名子中的稱謂合併剩下四種 分別為 Mr、Dr、Miss、Mrs、Master，而age的缺失值是利用這四種稱謂的平均值而填補")
data['Title1'] = data['name'].str.split(", ",expand=True)[1]
data['Title1'] = data['Title1'].str.split(". ",expand=True)[0]
print(data['Title1'].unique())

pd.crosstab(data['Title1'],data['sex']).T.style.background_gradient(cmap='summer_r')
# pd.crosstab(data['Title1'],data['survived']).T.style.background_gradient(cmap='summer_r')
```

```python
data['Title2'] = data['Title1'].replace(['Mlle','Mme','Ms','Major','Lady','th','Don','Col','Capt','Dona','Jonkheer','Rev','Sir'],
                                        ['Miss','Miss','Miss','Dr','Mrs','Mrs','Mr','Dr','Dr','Mrs','Mr','Mr','Dr'])

pd.crosstab(data['Title2'],data['sex']).T.style.background_gradient(cmap='summer_r')
```

## deal with age

```python

mask =  (data['age'].isnull()) & ( (data['name'].str.contains("Mrs")) | (data['name'].str.contains("Miss") |
                                                                         (data['name'].str.contains("Ms"))) )
mask2 = ( (data['name'].str.contains("Mrs")) | (data['name'].str.contains("Miss")) |
                                                                         (data['name'].str.contains("Ms")))
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())

mask =  (data['age'].isnull()) & ( (data['name'].str.contains("Mr")) | (data['name'].str.contains("Dr")))
mask2 = ( (data['name'].str.contains("Mr")) | (data['name'].str.contains("Dr")) )
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())  

mask =  (data['age'].isnull()) & ( data['name'].str.contains("Master"))
mask2 = ( (data['name'].str.contains("Master")))
data.loc[mask,'age'] = data.loc[mask,'age'].fillna(data.loc[mask2,'age'].mean())  

data.drop(['name','Title1'],axis=1,inplace = True) 
print("--------------------age已無缺失值--------------------")
Truthtable = data.isnull().any().tolist()

for i in range(len(Truthtable)):
    if(Truthtable[i] == True):
        print(data.columns.values[i],"此數據有缺失值")
```

## deal with 'fare'

```python
#根據生活經驗，年紀小的年紀很大的會比較便宜，其他都是正常票價，姑且假設沒有身心障礙等..
#將年齡分成[0,12)、[12~65),[65~對應三組的人票價取平均塞入缺失值
mask = data['age']<12
mask1 = (data['fare'].isnull() &  (data['age']<12))

data.loc[mask1,'fare'] = data.loc[mask1,'fare'].fillna(data.loc[mask,'fare'].mean())
mask = ((data['age']<65) & (12<=data['age'] ))
mask1 = (data['fare'].isnull() &  (12<=data['age']) & (data['age']<65))
data.loc[mask1,'fare'] = data.loc[mask1,'fare'].fillna(data.loc[mask,'fare'].mean())

mask = 65<=data['age']
mask1 = (data['fare'].isnull() &  (65<=data['age']))
data.loc[mask1,'fare'] = data.loc[mask1,'fare'].fillna(data.loc[mask,'fare'].mean())


print("--------------------fare已無缺失值--------------------")
Truthtable = data.isnull().any().tolist()

for i in range(len(Truthtable)):
    if(Truthtable[i] == True):
        print(data.columns.values[i],"此數據有缺失值")
```

## deal with others

```python
#觀察表格發現body cabin 缺失太多 其中body基本上都出現在沒搭到船的人 可以理解成屍體(?
#可以將body與boat一起討論 因此drop掉body與cabin
#有搭到救生船為1 否則為0
data.drop(['body','cabin'],axis=1,inplace = True) 
#有搭到救生船為1 否則為0
mask = data['boat'].isnull()
mask1 = data['boat'].notnull()
data.loc[mask,'boat'] = data.loc[mask,'boat'].fillna("0")  
data.loc[mask1,'boat']=1

print("--------------------缺失值--------------------")
Truthtable = data.isnull().any().tolist()

for i in range(len(Truthtable)):
    if(Truthtable[i] == True):
        print(data.columns.values[i],"此數據有缺失值")
```

```python
#我覺得home這影響不大drop掉
data.drop('home.dest',axis=1,inplace=True)
#因為embarked只有一個缺失值，隨便填一個進去
mask=data['embarked'].isnull()
data.loc[mask,'embarked']=data['embarked'].unique()[np.random.randint(3)]

print("--------------------缺失值--------------------")
Truthtable = data.isnull().any().tolist()

for i in range(len(Truthtable)):
    if(Truthtable[i] == True):
        print(data.columns.values[i],"此數據有缺失值")
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
data_train = data.drop('survived', axis=1)
data_train = data_train.head(873)
y_train = data['survived'].head(873)

print(data_train)

modelSVM = SVC()
modelSVM.fit(data_train, y_train)
data_pred = modelSVM.predict(data_train)

Score = accuracy_score(y_train, data_pred)
print('SVM accuracy: {:.3f}'.format(Score))
print (metrics.classification_report(y_train, data_pred) )


SVM_map = metrics.confusion_matrix(data_pred, y_train, [1,0])
sns.heatmap(SVM_map, annot=True, fmt='.2f',xticklabels = ["death", "survival"] , yticklabels = ["death", "survival"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('SVM')
plt.savefig('SVM')
```

```python
#使用Random Forest來預測train的survivied，並且畫出confusion_matrix

#n_estimator代表要使用多少CART樹（CART樹為使用GINI算法的決策樹）
rfc = RandomForestClassifier(n_estimators=100)

#從訓練組資料中建立隨機森林模型
rfc.fit(data_train, y_train)

rfc_pred = rfc.predict(data_train)
Score = accuracy_score(y_train, rfc_pred)
print('隨機森林 accuracy: {:.3f}'.format(Score))
print (metrics.classification_report(y_train, rfc_pred) )


rfc_map = metrics.confusion_matrix(rfc_pred, y_train, [1,0])
sns.heatmap(rfc_map, annot=True, fmt='.2f',xticklabels = ["death", "survival"] , yticklabels = ["death", "survival"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Random Forest')
plt.savefig('Random Forest')
```

```python
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn import metrics
from xgboost import XGBClassifier
import xgboost as xgb
from xgboost import plot_importance

#使用XgBoost來預測train的survivied，並且畫出confusion_matrix
params = {
    'booster': 'gbtree',
    'objective': 'multi:softmax',
    'num_class': 3,
    'gamma': 0.1,
    'max_depth': 6,
    'lambda': 2,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'min_child_weight': 3,
    'silent': 1,
    'eta': 0.1,
    'seed': 1000,
    'nthread': 4,
}

plst = list(params.items())
dtrain = xgb.DMatrix(data_train, y_train)
num_rounds = 500
model = xgb.train(plst, dtrain, num_rounds)

# 對測試集進行預測
dtest = xgb.DMatrix(data_train)
ans = model.predict(dtest)
Score = accuracy_score(y_train, ans)
print('xgboost accuracy: {:.3f}'.format(Score))
print (metrics.classification_report(y_train, ans) )

xgboot_map = metrics.confusion_matrix(ans, y_train, [1,0])
sns.heatmap(xgboot_map, annot=True, fmt='.2f',xticklabels = ["death", "survival"] , yticklabels = ["death", "survival"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('xgboot')
plt.savefig('xgboot')
```
