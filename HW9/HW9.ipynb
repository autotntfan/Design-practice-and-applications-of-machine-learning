```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression
from sklearn.linear_model import LinearRegression,Ridge,Lasso,RidgeCV,LassoCV
from sklearn import metrics

dataX = hr_df.drop(["sales","salary"],axis = 1)
dataY = hr_df["left"]

print(dataX)


X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.3) 
    
pls2 = PLSRegression(n_components=2)
# pls3 = Ridge(alpha=0.5)
# pls4 = Lasso(alpha=0.1)

model = pls2.fit(X_train, y_train)

Y_pred = pls2.predict(X_test)
       
mse = metrics.mean_squared_error(y_test, Y_pred)

print("Pred = ",Y_pred)
print("MSE = ",mse)
```

```python
alpha=10**np.linspace(-5,10,500)
RD_CV=RidgeCV(alphas=alpha,cv=10,scoring='r2')
RD_CV.fit(X_train,y_train)
pls3 = Ridge(alpha=RD_CV.alpha_)
model = pls3.fit(X_train, y_train)
Y_pred = pls3.predict(X_test)
mse = metrics.mean_squared_error(y_test, Y_pred)
print("Pred = ",Y_pred)
print("MSE = ",mse)
```

```python
pls4 = Lasso(alpha=0.5)
model = pls4.fit(X_train, y_train)
Y_pred = pls4.predict(X_test)
mse = metrics.mean_squared_error(y_test, Y_pred)
print("Pred = ",Y_pred)
print("MSE = ",mse)
```

```python
alpha=10**np.linspace(-5,10,500)
La_CV=LassoCV(alphas=alpha,cv=10)
La_CV.fit(X_train,y_train)
pls4 = Lasso(alpha=La_CV.alpha_)
model = pls4.fit(X_train, y_train)
Y_pred = pls4.predict(X_test)
mse = metrics.mean_squared_error(y_test, Y_pred)
print("Pred = ",Y_pred)
print("MSE = ",mse)
```

```python
#HW9 question1
wat=pd.read_csv('WAT.csv')
yied=pd.read_csv('Yield.csv')
wat.loc[:,'yield']=yied.iloc[:,1:2]
wat
```

```python
yield_corr=wat.corr().sort_values("yield",ascending=True)
yield_corr
```

```python
Name=yield_corr.index[-11:-1]
for name in Name:
    dataX=wat[name]
    dataY=wat['yield']
    X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.3) 
    pls2 = PLSRegression(n_components=1)
    model = pls2.fit(np.array(X_train).reshape(-1,1), y_train)
    Y_pred = pls2.predict(np.array(X_test).reshape(-1,1))
    plt.scatter(dataX,wat['yield'],color='b')
    plt.plot(X_test,Y_pred,color='r')
    plt.xlabel('wat')
    plt.ylabel('yield')
    plt.show()
```
